```python
import matplotlib.pyplot as plt
import numpy as np

def plot_single_model_performance(model_name, train_acc, train_loss, val_acc, val_loss):
    """
    Plots the performance (accuracy and log loss) of a single machine learning model.
    
    Parameters:
        model_name (str): Name of the model (e.g., 'SVM').
        train_acc (float): Training accuracy of the model.
        train_loss (float): Training log loss of the model.
        val_acc (float): Validation accuracy of the model.
        val_loss (float): Validation log loss of the model.
    """
    # Labels for the metrics
    metrics = ['Train Accuracy', 'Validation Accuracy', 'Train Log Loss', 'Validation Log Loss']
    
    # Scores for the metrics
    scores = [train_acc, val_acc, train_loss, val_loss]
    
    # Bar width for visibility
    width = 0.4
    
    # X-axis positions for the bars
    x = np.arange(len(metrics))

    # Create the plot
    fig, ax = plt.subplots(figsize=(8, 6))

    # Create bars for the scores
    ax.bar(x[:2], scores[:2], width, label='Accuracy', color='royalblue')  # Accuracy bars
    ax.bar(x[2:], scores[2:], width, label='Log Loss', color='darkorange')  # Log Loss bars

    # Add labels, title, and x-axis tick labels
    ax.set_ylabel('Scores', fontsize=14)
    ax.set_title(f'{model_name} Performance: Accuracy and Log Loss', fontsize=16)
    ax.set_xticks(x)
    ax.set_xticklabels(metrics, rotation=45, ha='right', fontsize=12)

    # Adding a legend
    ax.legend(loc='upper left', fontsize=12)

    # Adjust layout to make sure everything fits
    fig.tight_layout()

    # Show the plot
    plt.show()


# Example usage for multiple models:

# Model performance data
models_data = {
    'SVM': {
        'train_acc': 0.6697,
        'train_loss': 0.7692,
        'val_acc': 0.6700,
        'val_loss': 0.7441
    },
    'Random Forest': {
        'train_acc': 0.9996,
        'train_loss': 0.1434,
        'val_acc': 0.7250,
        'val_loss': 0.7103
    },
    'Logistic Regression': {
        'train_acc': 0.6620,
        'train_loss': 0.7683,
        'val_acc': 0.6790,
        'val_loss': 0.7409
    },
    'LightGBM': {
        'train_acc': 0.9784,
        'train_loss': 0.2254,
        'val_acc': 0.7260,
        'val_loss': 0.6667
    }
}

# Iterate through the models and plot each model's performance
for model_name, metrics in models_data.items():
    plot_single_model_performance(
        model_name, 
        metrics['train_acc'], 
        metrics['train_loss'], 
        metrics['val_acc'], 
        metrics['val_loss']
    )
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay

def plot_confusion_matrix(model_name, conf_matrix, class_names):
    """
    Plots a confusion matrix for a given model.

    Parameters:
        model_name (str): The name of the model (e.g., 'SVM').
        conf_matrix (array): The confusion matrix (2D array).
        class_names (list): List of class names.
    """
    # Create a figure for the confusion matrix plot
    fig, ax = plt.subplots(figsize=(8, 6))

    # Plotting the confusion matrix using seaborn heatmap
    sns.heatmap(conf_matrix, annot=True, fmt='g', cmap='Blues', xticklabels=class_names, yticklabels=class_names, ax=ax)

    # Set plot title and labels
    ax.set_title(f'{model_name} - Confusion Matrix', fontsize=16)
    ax.set_xlabel('Predicted Labels', fontsize=14)
    ax.set_ylabel('True Labels', fontsize=14)

    # Display the plot
    plt.tight_layout()
    plt.show()


# Example usage for multiple models:

# Model confusion matrices and class names
confusion_matrices = {
    'SVM': np.array([[583, 286, 70], [61, 124, 39], [53, 151, 633]]),
    'Random Forest': np.array([[756, 102, 81], [132, 42, 50], [136, 49, 652]]),
    'Logistic Regression': np.array([[610, 250, 79], [72, 106, 46], [59, 136, 642]]),
    'LightGBM': np.array([[725, 128, 86], [110, 61, 53], [96, 75, 666]])
}

# Class names for all models
class_names = ['negative', 'neutral', 'positive']

# Iterate through models and plot confusion matrices
for model_name, conf_matrix in confusion_matrices.items():
    plot_confusion_matrix(model_name, conf_matrix, class_names)
```
