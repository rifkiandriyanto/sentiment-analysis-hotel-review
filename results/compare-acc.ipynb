```python
import matplotlib.pyplot as plt
import numpy as np

# Data for accuracy from the classification reports
models = ['FastText-LSTM', 'Word2Vec-LSTM', 'Doc2Vec-LSTM']
accuracy = [0.83, 0.93, 0.96]

# Create a bar chart
x_pos = np.arange(len(models))

plt.figure(figsize=(8, 6))
plt.bar(x_pos, accuracy, color=['blue', 'green', 'red'], alpha=0.7)

# Add labels and title
plt.xlabel('Models', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('Accuracy Comparison of FastText, Word2Vec, and Doc2Vec with LSTM', fontsize=14)

# Add accuracy values on top of bars
for i in range(len(accuracy)):
    plt.text(x_pos[i], accuracy[i] + 0.01, f'{accuracy[i]:.2f}', ha='center', fontsize=12)

# Set the x-axis ticks
plt.xticks(x_pos, models)

# Display the plot
plt.ylim([0.75, 1.0])
plt.tight_layout()
plt.show()
```
