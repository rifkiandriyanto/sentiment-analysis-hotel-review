```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
labels = [
    'FastText-LSTM', 'Word2Vec-LSTM', 'Doc2Vec-LSTM', 
    'FastText-BiLSTM', 'Word2Vec-BiLSTM', 'Doc2Vec-BiLSTM', 
    'FastText-BiLSTM-Attention', 'Word2Vec-BiLSTM-Attention', 'Doc2Vec-BiLSTM-Attention'
]
accuracy = [0.83, 0.93, 0.96, 0.85, 0.95, 0.96, 0.85, 0.95, 0.96]
weighted_avg_recall = [0.83, 0.93, 0.96, 0.85, 0.95, 0.96, 0.85, 0.95, 0.96]
weighted_avg_f1_score = [0.85, 0.93, 0.96, 0.86, 0.95, 0.96, 0.86, 0.95, 0.96]
macro_avg_recall = [0.75, 0.90, 0.93, 0.79, 0.88, 0.92, 0.76, 0.92, 0.93]
macro_avg_f1_score = [0.73, 0.87, 0.93, 0.76, 0.90, 0.93, 0.74, 0.92, 0.93]

x = np.arange(len(labels))  # the label locations
width = 0.14  # the width of the bars

fig, ax = plt.subplots(figsize=(14,10))
rects1 = ax.bar(x - 2*width, accuracy, width, label='Accuracy')
rects2 = ax.bar(x - width, weighted_avg_recall, width, label='Weighted Avg Recall')
rects3 = ax.bar(x, weighted_avg_f1_score, width, label='Weighted Avg F1-Score')
rects4 = ax.bar(x + width, macro_avg_recall, width, label='Macro Avg Recall')
rects5 = ax.bar(x + 2*width, macro_avg_f1_score, width, label='Macro Avg F1-Score')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores', fontsize=16)
ax.set_title('Comparison of Model Performance', fontsize=18)
ax.set_xlabel('Models', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=18)

# Move the legend to inside the plot area, at the bottom left
ax.legend(loc='lower left', bbox_to_anchor=(0, -0.01), fontsize=16)

# Adjust the margins
fig.subplots_adjust(left=0.05, right=0.95)

fig.tight_layout()
plt.show()
```
