```python
import matplotlib.pyplot as plt
import numpy as np

def plot_comparison(labels, accuracy, weighted_avg_recall, weighted_avg_f1_score, macro_avg_recall, macro_avg_f1_score, title):
    x = np.arange(len(labels))  # the label locations
    width = 0.15  # the width of the bars

    fig, ax = plt.subplots(figsize=(10, 6))
    rects1 = ax.bar(x - 2*width, accuracy, width, label='Accuracy')
    rects2 = ax.bar(x - width, weighted_avg_recall, width, label='Weighted Avg Recall')
    rects3 = ax.bar(x, weighted_avg_f1_score, width, label='Weighted Avg F1-Score')
    rects4 = ax.bar(x + width, macro_avg_recall, width, label='Macro Avg Recall')
    rects5 = ax.bar(x + 2*width, macro_avg_f1_score, width, label='Macro Avg F1-Score')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_xlabel('Models')
    ax.set_ylabel('Scores')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, ha='right')

    # Move the legend to inside the plot area, at the bottom left
    ax.legend(loc='lower left', bbox_to_anchor=(0, -0.01))

    fig.tight_layout()
    plt.show()

# Full comparison
labels_full = [
    'FastText-LSTM', 'Word2Vec-LSTM', 'Doc2Vec-LSTM', 
    'FastText-BiLSTM', 'Word2Vec-BiLSTM', 'Doc2Vec-BiLSTM', 
    'FastText-BiLSTM-Attention', 'Word2Vec-BiLSTM-Attention', 'Doc2Vec-BiLSTM-Attention'
]
accuracy_full = [0.83, 0.93, 0.96, 0.85, 0.95, 0.96, 0.85, 0.95, 0.96]
weighted_avg_recall_full = [0.83, 0.93, 0.96, 0.85, 0.95, 0.96, 0.85, 0.95, 0.96]
weighted_avg_f1_score_full = [0.85, 0.93, 0.96, 0.86, 0.95, 0.96, 0.86, 0.95, 0.96]
macro_avg_recall_full = [0.75, 0.90, 0.93, 0.79, 0.88, 0.92, 0.76, 0.92, 0.93]
macro_avg_f1_score_full = [0.73, 0.87, 0.93, 0.76, 0.90, 0.93, 0.74, 0.92, 0.93]

plot_comparison(
    labels_full,
    accuracy_full,
    weighted_avg_recall_full,
    weighted_avg_f1_score_full,
    macro_avg_recall_full,
    macro_avg_f1_score_full,
    'Comparison of Model Performance'
)

# LSTM comparison
labels_lstm = ['FastText-LSTM', 'Word2Vec-LSTM', 'Doc2Vec-LSTM']
accuracy_lstm = [0.83, 0.93, 0.96]
weighted_avg_recall_lstm = [0.83, 0.93, 0.96]
weighted_avg_f1_score_lstm = [0.85, 0.93, 0.96]
macro_avg_recall_lstm = [0.75, 0.90, 0.93]
macro_avg_f1_score_lstm = [0.73, 0.87, 0.93]

plot_comparison(
    labels_lstm,
    accuracy_lstm,
    weighted_avg_recall_lstm,
    weighted_avg_f1_score_lstm,
    macro_avg_recall_lstm,
    macro_avg_f1_score_lstm,
    'Comparison of LSTM Model Performance'
)

# BiLSTM comparison
labels_bilstm = ['FastText-BiLSTM', 'Word2Vec-BiLSTM', 'Doc2Vec-BiLSTM']
accuracy_bilstm = [0.85, 0.95, 0.96]
weighted_avg_recall_bilstm = [0.85, 0.95, 0.96]
weighted_avg_f1_score_bilstm = [0.86, 0.95, 0.96]
macro_avg_recall_bilstm = [0.79, 0.88, 0.92]
macro_avg_f1_score_bilstm = [0.76, 0.90, 0.93]

plot_comparison(
    labels_bilstm,
    accuracy_bilstm,
    weighted_avg_recall_bilstm,
    weighted_avg_f1_score_bilstm,
    macro_avg_recall_bilstm,
    macro_avg_f1_score_bilstm,
    'Comparison of BiLSTM Model Performance'
)

# BiLSTM-Attention comparison
labels_bilstm_attention = ['FastText-BiLSTM-Attention', 'Word2Vec-BiLSTM-Attention', 'Doc2Vec-BiLSTM-Attention']
accuracy_bilstm_attention = [0.85, 0.95, 0.96]
weighted_avg_recall_bilstm_attention = [0.85, 0.95, 0.96]
weighted_avg_f1_score_bilstm_attention = [0.86, 0.95, 0.96]
macro_avg_recall_bilstm_attention = [0.76, 0.92, 0.93]
macro_avg_f1_score_bilstm_attention = [0.74, 0.92, 0.93]

plot_comparison(
    labels_bilstm_attention,
    accuracy_bilstm_attention,
    weighted_avg_recall_bilstm_attention,
    weighted_avg_f1_score_bilstm_attention,
    macro_avg_recall_bilstm_attention,
    macro_avg_f1_score_bilstm_attention,
    'Comparison of BiLSTM-Attention Model Performance'
)
```
