```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting focused on accuracy
labels = [
    'FastText-LSTM', 'Word2Vec-LSTM', 'Doc2Vec-LSTM', 
    'FastText-BiLSTM', 'Word2Vec-BiLSTM', 'Doc2Vec-BiLSTM', 
    'FastText-BiLSTM-Attention', 'Word2Vec-BiLSTM-Attention', 'Doc2Vec-BiLSTM-Attention'
]
accuracy = [0.83, 0.93, 0.96, 0.84, 0.95, 0.96, 0.85, 0.95, 0.96]

x = np.arange(len(labels))  # the label locations
width = 0.4  # the width of the bars

fig, ax = plt.subplots(figsize=(14, 8))
rects1 = ax.bar(x, accuracy, width, label='Accuracy', color='skyblue')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy', fontsize=16)
ax.set_title('Accuracy Comparison of Different Models', fontsize=18)
ax.set_xlabel('Models', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=14)

# Display accuracy values on top of bars
for i in range(len(accuracy)):
    ax.text(x[i], accuracy[i] + 0.01, f'{accuracy[i]:.2f}', ha='center', fontsize=12)

# Adjust the margins
fig.subplots_adjust(left=0.05, right=0.95)

fig.tight_layout()
plt.show()
```
