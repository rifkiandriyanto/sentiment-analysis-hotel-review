```python
import matplotlib.pyplot as plt

def plot_training_results(epochs, train_losses, train_accuracies, loss_title, accuracy_title):
    """
    Plot training results including training loss and training accuracy.
    Args:
    epochs (list): List of epochs.
    train_losses (list): List of training loss values.
    train_accuracies (list): List of training accuracy values.
    loss_title (str): Title for the training loss plot.
    accuracy_title (str): Title for the training accuracy plot.
    """
    plt.figure(figsize=(14, 7))

    # Plotting training loss
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_losses, label='Training Loss', color='b', marker='o')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(loss_title)
    plt.legend()

    # Plotting training accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_accuracies, label='Training Accuracy', color='g', marker='o')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(accuracy_title)
    plt.legend()

    plt.tight_layout()
    plt.show()

# Example usage:
epochs = list(range(1, 21))
```
