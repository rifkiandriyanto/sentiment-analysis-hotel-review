```python
import matplotlib.pyplot as plt

def plot_training_results(epochs, train_losses_folds, train_accuracies_folds, loss_title, accuracy_title, val_accuracy_title):
    """
    Plot training results including training loss, training accuracy, and validation accuracy.

    Args:
    epochs (list): List of epochs.
    train_losses_folds (list of lists): List of training loss values for each fold.
    train_accuracies_folds (list of lists): List of training accuracy values for each fold.
    loss_title (str): Title for the training loss plot.
    accuracy_title (str): Title for the training accuracy plot.
    val_accuracy_title (str): Title for the validation accuracy plot.
    """
    # Plotting training loss
    plt.figure(figsize=(14, 7))
    for i, train_losses in enumerate(train_losses_folds):
        plt.subplot(1, 2, 1)
        plt.plot(epochs, train_losses, label=f'Fold {i+1}')
    
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(loss_title)
    plt.legend()
    
    # Plotting training accuracy
    for i, train_accuracies in enumerate(train_accuracies_folds):
        plt.subplot(1, 2, 2)
        plt.plot(epochs, train_accuracies, label=f'Fold {i+1}')
    
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(accuracy_title)
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Example usage:
epochs = list(range(1, 21))
```
